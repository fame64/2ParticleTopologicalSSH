```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
J1=1
J2=10
k1arr=np.linspace(-np.pi/2, np.pi/2, 100)
k2=np.linspace(-np.pi/2, np.pi/2, 100)
q=[]
Klist=[]
Elist1=[]
Elist2=[]
Elist3=[]
Elist4=[]
Elist5=[]
Elist6=[]
plt.figure(figsize=[12,9])
for k1 in k1arr:
    K=k1+k2
    E1=np.sqrt(J1**2+J2**2+2*J1*J2*np.cos(2*k1))/J2
    E2=np.sqrt(J1**2+J2**2+2*J1*J2*np.cos(2*k2))/J2
    for i in range(100):
        q.append(k2[i]-k1)
        Klist.append(k2[i]+k1)
        Elist1.append(E1+E2[i])
        Elist2.append(-E1+E2[i])
        Elist3.append(E1-E2[i])
        Elist4.append(-E1-E2[i])
    plt.plot(K, E1+E2, color='cyan')
    plt.plot(K, -E1+E2, color='cyan')
    plt.plot(K, E1-E2, color='cyan')
    plt.plot(K, -E1-E2, color='cyan')
plt.plot(K, E1+E2, color='cyan', label='Scattering bands')
plt.plot(k2, E2, color='red', label='Bulk-Edge band')
plt.plot(k2, -E2, color='red')
plt.xlabel("Center of mass quasimomentum $K$", size=16)
plt.ylabel("$E/J_2$", size=16)
plt.legend()
plt.title("Energy bands of 2 particles SSH with Ï‰0=0, J1="+str(J1)+" J2="+str(J2), size=16)
plt.savefig("EnergyBandsScattering.png")
    #plt.plot(K, 1/J2*np.sqrt(2*(J1**2 + J2**2 + 2*J1*J2*np.cos(K)*np.cos(q))-0.5*np.sqrt((J1**2+J2**2+2*J1*J2*np.cos(K)*np.cos(q))**2)-4*(16*J1**2*J2*2*(np.sin(K))**2*(np.sin(q))**2)))
    #plt.plot(K, 1/J2*np.sqrt(2*(J1**2 + J2**2 + 2*J1*J2*np.cos(K)*np.cos(q))+0.5*np.sqrt((J1**2+J2**2+2*J1*J2*np.cos(K)*np.cos(q))**2)-4*(16*J1**2*J2*2*(np.sin(K))**2*(np.sin(q))**2)))
plt.show()

```
